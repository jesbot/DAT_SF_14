```python
# import pandas
import pandas as pd
# For plotting
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Read-in sub-sample bank data...
# Note: data is separated by semicolons, not commas, so specify when reading data
bankDF = pd.read_csv('bank.csv',sep=';')
```

```python
# 1. Use the file **bank.csv** to explore the dataset. Observe the features: Are they numbers? 
# Are they strings? Are they binary? Are they continuous?

# Look at the header of the data
bankDF.head(10)
```

```python
# Let's get some stat descriptions
bankDF.info()
```

So we have a grand total of 4,521 entries with 17 columns. 7 columns are numbers, (int64), and the remaining 10 are objects, most likely strings. 

From the columns, I think columns 'defualt, 'housing','loan', and 'y' are boolean, (yes/no). 'balance' can be continuous in nature, although it appersvalues have been rounded off to the nearest dollar. All other variables are descrete or string descriptors.

```python
# 2. Learn about label encoders at the following link and use what you learn to transform the features
# to numerical features. 
import numpy as np

# Create an encoded data frame
encodedDF = pd.DataFrame.copy(bankDF)

# Select the columns that need to be encoded
columns = np.array([1, 2, 3, 4, 6, 7, 8, 10, 15, 16])

# I used 'labelEncoder' instead of 'OneHotEncoder' 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Go through and encode all 'object'-type features
for col in bankDF:
    if bankDF[col].dtype == 'O':
        le.fit(bankDF[col])
        encodedDF[col] = le.transform(bankDF[col])
        
encodedDF.head(10)
```

```python
# 3. Build a decision tree model to predict whether a prospect will buy the product.

# Create labels for the data
featureColumns = encodedDF.columns[:-1]
targetColumn = encodedDF.columns[-1]

# Set data variabels
X = encodedDF[featureColumns]
y = encodedDF[targetColumn]

# Import and train data sets
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# Build a tree model
# import class, instantiate estimator, fit with training set
from sklearn.tree import DecisionTreeRegressor
treereg = DecisionTreeRegressor(random_state=12)
treereg.fit(X_train, y_train)

# Now do a standard predict to see if you have a tree
yPrediction = treereg.predict(X_test)
print yPrediction[1:50]
```

```python
# 4. Evaluate the accuracy of your decision tree model using cross validation.

# Import cross-validation
from sklearn.cross_validation import cross_val_score

# Calculate scores using cross-validation with 5 folds
scores = cross_val_score(treereg, X, y, cv=5, scoring='mean_squared_error')
print -scores
print "Mean squred error: {}".format(np.mean(np.sqrt(-scores)))
```

```python
# Calculate scores using cross-validation with 7 folds
scores = cross_val_score(treereg, X, y, cv=7, scoring='mean_squared_error')
print "Mean squred error: {}".format(np.mean(np.sqrt(-scores)))
```

```python
# Find the optimal depth of the tree from 3 to 10
optScores = np.zeros([1,len(range(3,11))])
treeDepth = np.zeros([1,len(range(3,11))])
index = 0
for i in range(3,11):
    treeregOpt = DecisionTreeRegressor(max_depth=i, random_state=12)
    treeregOpt.fit(X_train, y_train)
    scores = cross_val_score(treeregOpt, X, y, cv=5, scoring='mean_squared_error')
    optScores[0,index] = np.mean(np.sqrt(-scores))
    treeDepth[0,index] = i
    index += 1

print optScores
print treeDepth
# Plot results to find optimal depth
plt.plot(treeDepth[0,:],optScores[0,:],linewidth=3.0)
plt.title('Optimized Tree Depth')
plt.ylabel('MSE')
plt.xlabel('Depth of Tree')
```

So it appers an optimal tree in terms of 5-fold CV is 4. Let's just plot to see what that tree looks like.

```python
# Create a visual tree
treeregOptPlot = DecisionTreeRegressor(max_depth=4, random_state=12)
treeregOptPlot.fit(X_train, y_train)
from sklearn.tree import export_graphviz
with open("bankDF_subsample.dot", 'wb') as f:
    f = export_graphviz(treeregOptPlot, out_file=f, \
                        feature_names=featureColumns)
# Below is the resulting tree
```
